```python
import numpy as np
import pandas as pd
import cv2
from time import sleep
import warnings
warnings.filterwarnings('ignore')
```

```python
# Parameters for vehicle detection
min_width_rect = 80    # Minimum width of vehicle rectangle
min_height_rect = 80   # Minimum height of vehicle rectangle
offset = 6             # Tolerance around counting line
delay = 60            # Delay between detections
carros = 0            # Vehicle counter
count_line_pos = 550  # Y-position of counting line
detec = []            # List to store detected vehicle centers


## Center Point Calculator:
def central_handle(x,y,w,h):
    x1 = int(w/2)
    y1 = int(h/2)
    cx = x + x1   # Center x-coordinate
    cy = y + y1   # Center y-coordinate
    return cx, cy

# Enable web camera
cap = cv2.VideoCapture('video.mp4')
#algo = cv2.createBackgroundSubtractorMOG2()
algo = cv2.createBackgroundSubtractorKNN()

while True:
    # Read and process frame
    ret, frame1 = cap.read()
    
    # Image processing
    gray = cv2.cvtColor(frame1, cv2.COLOR_BGR2GRAY)
    blur = cv2.GaussianBlur(gray, (3,3), 5)
    
    #  Apply Background subtraction
    img_sub = algo.apply(blur)

    # Morphological operations
    dilat = cv2.dilate(img_sub, np.ones((5,5)))
    kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5,5))
    dilatada = cv2.morphologyEx(dilat, cv2.MORPH_CLOSE, kernel)
    dilatada = cv2.morphologyEx(dilatada, cv2.MORPH_CLOSE, kernel)
    counterShape, h = cv2.findContours(dilatada, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    cv2.line(frame1, (25,count_line_pos), (1200,count_line_pos),(0,0,255), 3)

    # Process each contour
    for (i,c) in enumerate(counterShape):
        (x,y,w,h) = cv2.boundingRect(c)
        validar_contorno = (w >= min_width_rect) and (h >=min_height_rect)

        if not validar_contorno:
            continue
        cv2.rectangle(frame1, (x,y), (x+w, y+h), (0,255,0), 2)
        cv2.putText(frame1, "Vehicle" + str(carros), (x, y-20), cv2.FONT_HERSHEY_TRIPLEX, 1, (255,0,255),2)
        center = central_handle(x,y,w,h)
        detec.append(center)
        cv2.circle(frame1, center, 4,(0,0,255),-1)

        # Loop Function - checking if the vehicle crossed the line
        for (x,y) in detec:
            if y<(count_line_pos+offset) and y>(count_line_pos-offset):
                carros +=1
                 # Draw counting line
                cv2.line(frame1, (25,count_line_pos), (1200,count_line_pos),(0,0,255), 3)
                detec.remove((x,y))
                print("Car is detecting :" +str(carros))
    # Add vehicle label
    cv2.putText(frame1, "Vehicle" + str(carros), (x, y-20), cv2.FONT_HERSHEY_TRIPLEX, 1, (255,0,255),2)

    cv2.imshow("Original Video", frame1)

    if cv2.waitKey(1)==13:
        break
cap.release()
cv2.destroyAllWindows()
```
